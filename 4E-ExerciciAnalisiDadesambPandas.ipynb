```python
import pandas as pd
```

```python
notes = pd.DataFrame([
    [6.5,6.8,5.8,7.9,8.8,'Barcelona'], 
    [6.8,7.0,6.5,8.9,9.2,'Girona'],
    [7.8,6.2,6.5,4.1,8.0,'Tarragona'], 
    [6.9,7.2,4.7,8.8,6.2,'Lleida']],
    columns=['Història','Matemàtiques','Física','Anglès','Biologia','Ciutat'],
    index=['Joan','Laura','Enric','Georgina',])

notes
```

```python
mitjanes = []
desviacio = []

for i in range(4):
    mitjana = notes.iloc[i, 0:5].mean()
    desv = notes.iloc[i, 0:5].std()
    mitjanes.append(mitjana)
    desviacio.append(desv)

notes['Mitjanes'] = mitjanes
notes['Desviacio'] = desviacio

notes
```
